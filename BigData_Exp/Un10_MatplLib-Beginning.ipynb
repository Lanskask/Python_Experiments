# Matplotlib Beginning

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

# Functional style ploting

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, '-r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Plot 1')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'y')
```

# OO Style Ploting

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # - size of canvas
axes.plot(x,y) # - should be in the same cell, not in the other
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title OO 2')
```

```python
fig = plt.figure()
axes2 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes3 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes2.plot(x,y)
axes2.set_title('Title OUTS')
axes2.set_xlabel('X Out')
axes2.set_ylabel('Y Out')

axes3.plot(y,x)
axes3.set_title('Title INS')
axes3.set_xlabel('X Ins')
axes3.set_ylabel('Y Ins')
```

## Making subplots

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes: 
    current_ax.plot(x,y)
axes[0].plot(y,x)
axes[1].plot(x,y**(1/2))
```

# Figure Siza and DPI

```python
fig = plt.figure(figsize=(6,2),dpi=300)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure(figsize=(6,2),dpi=300)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Square')
ax.plot(x,x**3, label='X Cube')

ax.legend(loc=(0.1,0.1)) # or for 0 to 10 includes
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# Plot Apearence 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,
        color='#FF0000', 
        linewidth=3, 
        alpha=0.3,
        linestyle='steps', # --, :, steps, -. # ls - instead of full name
        marker='o' # o, +, *, 1 ,2,3,',', s   so on
        , markersize=20
        , markerfacecolor='y'
        , markeredgewidth=3
       ) 
# color could be like 'orange', 'o', rgb code
# it possible lw instead of linewidth
# alpha - add a little tranperency - прозрачность
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,
        color='purple', 
        lw=3, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
