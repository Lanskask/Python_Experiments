# Matplotlib Beginning

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

# Functional style ploting

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, '-r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Plot 1')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'y')
```

# OO Style Ploting

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # - size of canvas
axes.plot(x,y) # - should be in the same cell, not in the other
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title OO 2')
```

```python
fig = plt.figure()
axes2 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes3 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes2.plot(x,y)
axes2.set_title('Title OUTS')
axes2.set_xlabel('X Out')
axes2.set_ylabel('Y Out')

axes3.plot(y,x)
axes3.set_title('Title INS')
axes3.set_xlabel('X Ins')
axes3.set_ylabel('Y Ins')
```

## Making subplots

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes: 
    current_ax.plot(x,y)
axes[0].plot(y,x)
axes[1].plot(x,y**(1/2))
```

```python
fig = plt.figure(figsize=(8,2),dpi=300)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```
