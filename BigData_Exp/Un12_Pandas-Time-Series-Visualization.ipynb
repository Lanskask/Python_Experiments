```python
fig,ax = plt.subplots()
ax.plot_date(idx, stock, '-')

# Locating 
ax.xaxis.set_major_locator(dates.MonthLocator())
# Formating
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx, stock, '-')

## MAJOR Locator
# Locating 
ax.xaxis.set_major_locator(dates.MonthLocator())
# Formating
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

## MINOR Locator
ax.xaxis.set_minor_locator(dates.WeekdayLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx, stock, '-')

## MAJOR Locator
# Locating 
ax.xaxis.set_major_locator(dates.MonthLocator())
# Formating
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

## MINOR Locator
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%a'))
# byweekday=0 - in WeekdayLocator to see Mondays on graphic
# '%a' for days of week

fig.autofmt_xdate()
plt.tight_layout()
```
